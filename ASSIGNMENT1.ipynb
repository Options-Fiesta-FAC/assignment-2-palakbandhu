```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def spot_price (K,P):
    list1=[P]*101
    list2=[K]*101
    data={"spot_price":list(range(K-50,K+51)),"premium":list1,"strike_price":list2}
    df=pd.DataFrame(data)
    
    df["BuyCall"] = np.maximum(0, df["spot_price"] - df["strike_price"]) - df["premium"]
    df["SellCall"] = -np.maximum(0, df["spot_price"] - K) + P
    df["BuyPut"] = np.maximum(0, K - df["spot_price"]) - P
    df["SellPut"] = P - np.maximum(0, K - df["spot_price"])
    import matplotlib.pyplot as plt



    fig, axs = plt.subplots(4,1, figsize=(12, 15))  # 2 rows, 2 columns

    # First subplot: Buy Call
    axs[0].plot(df['spot_price'], df['BuyCall'], color='blue')
    axs[0].set_title('Buy Call Payoff')
    axs[0].set_xlabel('Spot Price')
    axs[0].set_ylabel('Payoff')
    axs[0].axhline(0, color='black', linewidth=1)
    axs[0].axvline(100, color='grey', linestyle='--')  # Example: strike price at 100

    # Second subplot: Sell Call
    axs[ 1].plot(df['spot_price'], df['SellCall'], color='orange')
    axs[ 1].set_title('Sell Call Payoff')
    axs[1].set_xlabel('Spot Price')
    axs[1].set_ylabel('Payoff')
    axs[ 1].axhline(0, color='black', linewidth=1)
    axs[1].axvline(100, color='grey', linestyle='--')

    # Third subplot: Buy Put
    axs[ 2].plot(df['spot_price'], df['BuyPut'], color='red')
    axs[2].set_title('Buy Put Payoff')
    axs[2].set_xlabel('Spot Price')
    axs[2].set_ylabel('Payoff')
    axs[2].axhline(0, color='black', linewidth=1)
    axs[ 2].axvline(100, color='grey', linestyle='--')

    # Fourth subplot: Sell Put
    axs[ 3].plot(df['spot_price'], df['SellPut'], color='green')
    axs[3].set_title('Sell Put Payoff')
    axs[3].set_xlabel('Spot Price')
    axs[3].set_ylabel('Payoff')
    axs[3].axhline(0, color='black', linewidth=1)
    axs[3].axvline(100, color='grey', linestyle='--')

    plt.tight_layout()
    plt.show()

    return df

```

```python
df=spot_price(100,10)
print(df.to_string())

```
